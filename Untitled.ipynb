```python
%%time
from imblearn.under_sampling import TomekLinks
from imblearn.over_sampling import SMOTE
from imblearn.combine import SMOTETomek
tm = SMOTETomek(random_state=9786)
x,y = tm.fit_sample(train,label)
x = pd.DataFrame(x)
```

```python
def splitting(m, train= train, label = label, test = test, random_state = 400, split = 3):
    a = list()
    b = list()
    sss=StratifiedShuffleSplit(random_state= random_state, test_size=0.2,  n_splits=split)
    pred_tr = pd.DataFrame(np.zeros(shape = (train.shape[0], 1)))
    pred_te = pd.DataFrame(np.zeros(shape = (test.shape[0], split)))

    for i, (tr_index, te_index) in enumerate(sss.split(train,label)):
        xtrain, xtest = train.loc[tr_index,:], train.loc[te_index,:]
        ytrain, ytest = label[tr_index], label[te_index]   
        m.fit(xtrain,ytrain)
        p = m.predict(xtest)
        pred_tr.loc[tr_index]
        pred_te.iloc[:,i] = m.predict(test)
        
        print(f"Training : {f1_score(ytrain,m.predict(xtrain))}   Testing : {f1_score(ytest,m.predict(xtest))}")        
        a.append(f1_score(ytrain,m.predict(xtrain)))
        b.append(f1_score(ytest,m.predict(xtest)))
            
    print(pd.Series(m.feature_importances_, index = train.columns).sort_values(ascending=True).plot.barh())
    print(f'\nTraining Mean: {np.mean(a)} \nTesting Mean: {np.mean(b)}')
    #print(f'\nAUC SCORE: {roc_auc_score(pred, label)}')
    #print(f'Precision: {precision_score(pred, label)}')
    #print(f'Recall: {recall_score(pred, label)}')
    #print(f'F1 Score: {f1_score(pred, label)}')
    return(pred_te)
    #return((pd.DataFrame(m.predict(test))))
```
